```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator
```

```python
df = pd.read_csv('/Features_For_Traditional_ML_Techniques.csv')
```

```python
# Display basic information about the dataset
print(df.info())
```

```python
# Distribution of the data set
print(df.describe())
```

```python
# Set the style for seaborn
sns.set(style="whitegrid")
```

```python
# Distribution of Target Variable (majority_target)
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='majority_target')
plt.title('Distribution of Target Variable (majority_target)')
plt.xlabel('Truth Value (1=True, 0=False)')
plt.ylabel('Count')
plt.show()
```

```python
#Mean Follower and Friend count by Majority Target

# Calculate mean counts for each majority_target
followers_mean = df.groupby('majority_target')['followers_count'].mean()
friends_mean = df.groupby('majority_target')['friends_count'].mean()

# Create a figure with 2 subplots side by side
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Plot for Mean Followers Count
axes[0].bar(followers_mean.index, followers_mean.values, color='b', alpha=0.7)
axes[0].set_title('Mean Followers Count by Majority Target')
axes[0].set_xlabel('Majority Target (0=False, 1=True)')
axes[0].set_ylabel('Mean Followers Count')
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['False', 'True'])
axes[0].grid(False)  # Remove grid for a cleaner look

# Plot for Mean Friends Count
axes[1].bar(friends_mean.index, friends_mean.values, color='g', alpha=0.7)
axes[1].set_title('Mean Friends Count by Majority Target')
axes[1].set_xlabel('Majority Target (0=False, 1=True)')
axes[1].set_ylabel('Mean Friends Count')
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['False', 'True'])
axes[1].grid(False)  # Remove grid for a cleaner look

# Adjust layout for better spacing
plt.tight_layout()
plt.show()

```

```python
# Distribution of Pronoun Count, Adjectives count, Determiner Count,Conjunction Count,Dots Count

# Set the style for seaborn
sns.set(style='whitegrid')

# Create a figure with subplots
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(12, 25))  # Adjusted size for remaining graphs

# Plot for Pronoun Count
sns.histplot(df['pronouns'], bins=30, color='skyblue', ax=axes[0, 0], stat="count")
axes[0, 0].set_title('Distribution of Pronoun Count')
axes[0, 0].set_xlabel('Number of Pronouns')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].yaxis.set_major_locator(MaxNLocator(integer=True))
axes[0, 0].xaxis.set_major_locator(MaxNLocator(integer=True))

# Plot for Adjective Count
sns.histplot(df['adjectives'], bins=30, color='orange', ax=axes[1, 0], stat="count")
axes[1, 0].set_title('Distribution of Adjective Count')
axes[1, 0].set_xlabel('Number of Adjectives')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].yaxis.set_major_locator(MaxNLocator(integer=True))
axes[1, 0].xaxis.set_major_locator(MaxNLocator(integer=True))

# Plot for Determiner Count
sns.histplot(df['determiners'], bins=30, color='green', ax=axes[2, 0], stat="count")
axes[2, 0].set_title('Distribution of Determiner Count')
axes[2, 0].set_xlabel('Number of Determiners')
axes[2, 0].set_ylabel('Frequency')
axes[2, 0].yaxis.set_major_locator(MaxNLocator(integer=True))
axes[2, 0].xaxis.set_major_locator(MaxNLocator(integer=True))

# Plot for Conjunction Count
sns.histplot(df['conjunctions'], bins=30, color='red', ax=axes[3, 0], stat="count")
axes[3, 0].set_title('Distribution of Conjunction Count')
axes[3, 0].set_xlabel('Number of Conjunctions')
axes[3, 0].set_ylabel('Frequency')
axes[3, 0].yaxis.set_major_locator(MaxNLocator(integer=True))
axes[3, 0].xaxis.set_major_locator(MaxNLocator(integer=True))

# Plot for Dots Count
sns.histplot(df['dots'], bins=30, color='blue', ax=axes[4, 0], stat="count")
axes[4, 0].set_title('Distribution of Dots Count')
axes[4, 0].set_xlabel('Number of Dots')
axes[4, 0].set_ylabel('Frequency')
axes[4, 0].yaxis.set_major_locator(MaxNLocator(integer=True))
axes[4, 0].xaxis.set_major_locator(MaxNLocator(integer=True))

# Hide the empty subplots
axes[0, 1].axis('off')
axes[1, 1].axis('off')
axes[2, 1].axis('off')
axes[3, 1].axis('off')
axes[4, 1].axis('off')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()

```

```python
# Present and Past Tense Verb Counts by Majority Target

# Set up the figure and axes for two side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot for Present Tense Verbs by Majority Target
sns.boxplot(data=df, x='majority_target', y='present_verbs', palette="magma", ax=axes[0])
axes[0].set_title('Present Tense Verbs by Majority Target')
axes[0].set_xlabel('Majority Target (1=True, 0=False)')
axes[0].set_ylabel('Present Tense Verb Count')

# Plot for Past Tense Verbs by Majority Target
sns.boxplot(data=df, x='majority_target', y='past_verbs', palette="viridis", ax=axes[1])
axes[1].set_title('Past Tense Verbs by Majority Target')
axes[1].set_xlabel('Majority Target (1=True, 0=False)')
axes[1].set_ylabel('Past Tense Verb Count')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
#Box plot of majority target agaisnt credibility score

# Round the 'cred' column to 1 decimal place
df['cred_rounded'] = df['cred'].round(1)

# Set the style for seaborn
sns.set(style='whitegrid')

# Create a box plot to show the relationship between majority_target and rounded credibility score
plt.figure(figsize=(8, 6))
sns.boxplot(x='majority_target', y='cred_rounded', data=df, palette='Set2')

# Add titles and labels
plt.title('Credibility Score by Majority Target')
plt.xlabel('Majority Target (0=False, 1=True)')
plt.ylabel('Credibility Score (Rounded to 1 Decimal Place)')
plt.xticks(ticks=[0, 1], labels=['False', 'True'])
plt.grid(axis='y')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
# Line graph of credibility of tweet against mean of replies and Retweets

# Round the credibility score to 1 decimal place
df['cred_rounded'] = df['cred'].round(1)

# Calculate the mean replies and retweets for each rounded cred value
mean_replies = df.groupby('cred_rounded')['replies'].mean()
mean_retweets = df.groupby('cred_rounded')['retweets'].mean()

# Create a figure for the line plots
plt.figure(figsize=(12, 6))

# Plot mean replies
plt.plot(mean_replies.index, mean_replies.values, label='Mean Replies', marker='o', color='blue')

# Plot mean retweets
plt.plot(mean_retweets.index, mean_retweets.values, label='Mean Retweets', marker='o', color='orange')

# Adding titles and labels
plt.title('Mean Replies and Retweets by Credibility Score (Rounded to 1 DP)')
plt.xlabel('Credibility Score (Rounded to 1 DP)')
plt.ylabel('Mean Counts')
plt.xticks(mean_replies.index)  # Set x-ticks to the index of mean replies
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
